# How to write CSV in pandas

```python
import pandas as pd
```

```python
df=pd.read_csv('D:\R_practice\house_R1.csv')
df
```

```python
type(df)
```

```python
#columns
df.columns
```

```python
df=pd.read_csv('D:\R_practice\house_R1.csv',nrows=4)
df
```

```python
#using columns
df=pd.read_csv('D:\R_practice\house_R1.csv',usecols=[1])
df
```

```python
# using skiprows
df=pd.read_csv('D:\R_practice\house_R1.csv',skiprows = [1,2,3])
df
```

```python
# index_col
df=pd.read_csv('D:\R_practice\house_R1.csv',index_col='Price')
df
```

```python
# 
df=pd.read_csv('D:\R_practice\house_R1.csv',header=[1])
df
```

```python
# 
df=pd.read_csv('D:\R_practice\house_R1.csv',header=None)
df
```

```python
# prefix
df=pd.read_csv('D:\R_practice\house_R1.csv',header=None, prefix = 'cols')
df
```

```python
# name of colums
df=pd.read_csv('D:\R_practice\house_R1.csv',names=[1,2])
df
```
